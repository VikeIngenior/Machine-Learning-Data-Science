```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity='all'

from sklearn import set_config
set_config(transform_output='pandas')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('data/feature-engine/credit.csv')
data['A16']=data['A16'].replace(to_replace={'+':1,'-':0})
num_vars = data.select_dtypes(include=[np.float64, np.int64]).columns
cat_vars = data.select_dtypes(exclude=[np.float64, np.int64]).columns
data[num_vars]=data[num_vars].fillna(0)
data[cat_vars]=data[cat_vars].fillna('Missing')
data.isna().sum()
data.to_csv('data/feature-engine/creditApprovalUCI-CategoricalAnalysis.csv',index=False)
```

```python
data=pd.read_csv('data/feature-engine/creditApprovalUCI-CategoricalAnalysis.csv')
x=data.drop('A16', axis=1)
y=data[['A16']]

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3, random_state=0)
```

```python
#sklearn
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder

#feature_engine
#from feature_engine.encoding import OneHotEncoder
```

```python
encoder = OneHotEncoder(categories='auto',
                        drop='first',
                        sparse_output=False)
encoder.fit(X_train[cat_vars])
X_train_enc = encoder.transform(X_train[cat_vars])
X_test_enc = encoder.transform(X_test[cat_vars])
X_train_enc.head()
```

```python
encoder = OrdinalEncoder()

encoder.fit(X_train[cat_vars])
X_train_enc=encoder.transform(X_train[cat_vars])
X_test_enc=encoder.transform(X_test[cat_vars])
X_train_enc.head()
```

```python
y_ordinal = np.array(['medium','large','small','medium'])
y_ordinal = y_ordinal[:,np.newaxis]
oe_1 = OrdinalEncoder(categories='auto')
oe_2 = OrdinalEncoder(categories=[['small','medium','large']])
oe_1.fit_transform(y_ordinal)
oe_2.fit_transform(y_ordinal)
```

```python
y= np.array(['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot'])
le = LabelEncoder()
le.fit_transform(y)
le.inverse_transform([0,2,1,2])
```

```python
from sklearn.preprocessing import TargetEncoder

encoder = TargetEncoder(target_type='binary')
encoder.fit(X_train[cat_vars], y_train.to_numpy().flatten())

x_train_enc = encoder.transform(X_train[cat_vars])
x_test_enc = encoder.transform(X_test[cat_vars])

X_train_enc.head()

X_train.head()
```
