# Join Operations

```python
a = np.random.randint(1,30,size=(5,3))
df1 = pd.DataFrame(a, columns = ["var1","var2","var3"])
df2 = df1 * 3
df1
```

```python
df2
```

```python
pd.concat([df1,df2]) #The issue here is indexing. That leads to ignore_index arguement
```

```python
pd.concat([df1,df2], ignore_index = True)
```

Now indexes are in order

```python
#Now I will fix the issue of variable(column) names
```

```python
df1.columns
```

```python
df2.columns
```

```python
df2.columns = ["var1","var2","deg3"]
```

```python
pd.concat([df1,df2])
```

Variable names don't fit. So I only want columns which intersect.

```python
pd.concat([df1,df2], join="inner", ignore_index=True)
```

```python
# join_axes arguement. This allows me get every variable names of the dataframe I want when I do the concat operation
#pd.concat([df1,df2], join_axes = [df1.columns])

# I found out that join_axes parameter has been removed from pd.concat. So here is another way to do the operation I want:

df2_reindexed = df2.reindex(columns=df1.columns)
```

df2 is reindexed to fit df1

```python
# Concatenate df1 and df2_reindexed
pd.concat([df1, df2_reindexed], ignore_index=True, sort=False)
```

# Mergings

One-to-one merge

```python
df1 = pd.DataFrame({'employees':['John','Jack','Linda','Susan'],
                    'group':['Accountant','Engineer','Engineer','HR']})
df1
```

```python
df2 = pd.DataFrame({'employees':['Linda','John','Jack','Susan'],
                   'start date':[2010,2009,2005,2011]})
df2
```

```python
pd.merge(df1,df2, on = 'employees') # var name other than 'employees' wouldn't work because it is the only common var name
                                    # between two dataframes
```

Many-to-one merge

```python
df3 = pd.merge(df1,df2)
```

```python
df3
```

```python
df4=pd.DataFrame({'group':['Accountant','Engineer','HR'],
                 'manager':['Curtis','Randy','Eva']})
df4
```

```python
pd.merge(df3,df4)
```

Many-to-many

```python
df5 = pd.DataFrame({'group':['Accountant','Accountant','Engineer','Engineer','HR','HR'],
                   'skill':['math','excel','programming','OS','excel','management']})
df5
```

```python
pd.merge(df1,df5)
```

```python
pd.merge(df5,pd.merge(df3,df4))
```

```python
# Here, I've wanted to change the index of columns 'group' and 'employees', but it seems it was pointless... 
df6 = pd.merge(df5,pd.merge(df3,df4))
df6_sutunlari = df6.columns.tolist()
df6_sutunlari[0],df6_sutunlari[2]=df6_sutunlari[2],df6_sutunlari[0]
df6 = df6[df6_sutunlari]
df6
```

Aggregate
<h6>filter, transform, apply</h6>

```python
df = pd.DataFrame({"groups":['A','B','C','A','B','C'],
                  "var1":[423,522,45,87,34,91],
                  "var2":[23,51,85,869,86,98]}, columns=["groups","var1","var2"])
df
```

```python
df.groupby("groups").min()
```

```python
df.groupby("groups").aggregate(["min",max,"mean",np.median])
```

```python
df.groupby("groups").aggregate({"var1":"min", "var2":"max"})
```

```python
def to_filter(x):
    return x["var1"].std() >300
```

```python
df.groupby("groups").filter(to_filter)
```

```python
df.drop("groups",axis=1,inplace=True)
df.transform(lambda x: (x/x.mean()) / x.std())
```

```python
df.apply(np.mean)
```
